```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OrdinalEncoder

import lux
```

```python
data = pd.read_csv('/content/train_dataset (1).csv')
data.head()
```

```python
data.info()
```

```python
fig, axes = plt.subplots(2,2, figsize=(15,8))

sns.countplot(x='Gender', hue='satisfaction', palette='viridis', data=data, ax=axes[0,0])
sns.countplot(x='Class', hue='satisfaction', palette='viridis', data=data, ax=axes[0,1])
sns.countplot(x='Customer Type', hue='satisfaction', palette='viridis', data=data, ax=axes[1, 0])
sns.countplot(x='Type of Travel', hue='satisfaction', data=data, ax=axes[1, 1] )
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(18, 8))
sns.countplot(x='Age', hue='satisfaction' , data=data)
plt.grid()
plt.show()
```
